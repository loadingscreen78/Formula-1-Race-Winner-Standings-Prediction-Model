Select Features and Split Data

```python
df.fillna(0, inplace=True)

# UPDATED FEATURE LIST
features = [
    'grid_position',
    'circuitId',
    'year',
    'driver_points_before_race',
    'driver_standings_before_race',
    'constructor_points_before_race',
    'constructor_standings_before_race'

]
target = 'final_position'

final_df = df[features + [target]]

train_df = final_df[final_df['year'] < 2023]
test_df = final_df[final_df['year'] >= 2023]

X_train = train_df.drop(columns=[target, 'year'])
y_train = train_df[target]
X_test = test_df.drop(columns=[target, 'year'])
y_test = test_df[target]

features = X_train.columns.tolist()

print(f"Updated features for the model: {features}")
print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```

Training

```python
y_train_adj = y_train - 1
y_test_adj = y_test - 1

model = xgb.XGBClassifier(
    objective='multi:softprob',
    num_class=35,


    n_estimators=1000,          # Increase estimators but rely on early stopping
    learning_rate=0.01,        # Very low learning rate
    max_depth=3,               # Very shallow trees
    subsample=0.6,
    colsample_bytree=0.6,
    gamma=0.2,

    use_label_encoder=False,
    eval_metric='mlogloss',
    early_stopping_rounds=75   # Increase patience for the low learning rate
)

model.fit(
    X_train, y_train_adj,
    eval_set=[(X_test, y_test_adj)],
    verbose=False
)

print("✅ Model re-trained with improved features and stronger regularization.")
```

```python
predictions = model.predict(X_test)
accuracy = np.mean(predictions == y_test_adj)
print(f"Model prediction accuracy on test set: {accuracy * 100:.2f}%")

plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots(figsize=(10, 8))
xgb.plot_importance(model, ax=ax, max_num_features=10, importance_type='weight', height=0.8)
plt.title('XGBoost Feature Importance', fontsize=16)
plt.xlabel('Importance (Weight)', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Ensure you have the actual and predicted values from the previous steps
# y_test_adj are the true positions (0-indexed)
# predictions are the model's predicted positions (0-indexed)

# Generate the confusion matrix
cm = confusion_matrix(y_test_adj, predictions)

# To make it readable, let's display only the top 10 positions
display_labels = [str(i) for i in range(1, 11)]

fig, ax = plt.subplots(figsize=(10, 10))
disp = ConfusionMatrixDisplay(confusion_matrix=cm[:10, :10], display_labels=display_labels)

disp.plot(ax=ax, cmap='Blues', values_format='d')
ax.set_title('Confusion Matrix (Top 10 Positions)', fontsize=16)
ax.set_xlabel('Predicted Position', fontsize=12)
ax.set_ylabel('Actual Position', fontsize=12)
plt.show()
```
